Вариант задания № 23
Описание проекта:
Описание задачи: Создать базовый класс Task в C#, который будет представлять задачи внутри проекта. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. Требования к базовому классу Task: • Атрибуты: ID задачи (TaskId), Название задачи (TaskName), Приоритет задачи (Priority). • Методы: o MarkAsComplete(): метод для отметки задачи как выполненной. o GetTaskDetails(): метод для получения деталей задачи. o ReassignTo(): метод для переназначения задачи другому члену команды. Требования к производным классам:

ДелегатскаяЗадача (DelegateTask): Должна содержать дополнительные атрибуты, такие как Дата выполнения (DueDate). Метод MarkAsComplete() должен быть переопределен для включения даты выполнения в сообщение о завершении задачи.
КоманднаяЗадача (TeamTask): Должна содержать дополнительные атрибуты, такие как Команда (TeamName). Метод ReassignTo() должен быть переопределен для указания нового члена команды, которому будет переназначена задача.
ИсследовательскаяЗадача (ResearchTask) (если требуется третий класс): Должна содержать дополнительные атрибуты, такие как Исходные данные (DataSource). Метод GetTaskDetails() должен быть переопределен для отображения источников данных, используемых в задаче, вместе с другими деталями задачи.


#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```

using System;

public class Task
{
    public int TaskId { get; set; }
    public string TaskName { get; set; }
    public string Priority { get; set; }
    public string Status { get; set; }
    public string Assignee { get; set; }
    public DateTime CreatedDate { get; set; }
    
    public Task(int taskId, string taskName, string priority, string assignee)
    {
        TaskId = taskId;
        TaskName = taskName;
        Priority = priority;
        Assignee = assignee;
        Status = "Новая";
        CreatedDate = DateTime.Now;
    }
    
    public virtual void MarkAsComplete()
    {
        Status = "Выполнена";
        Console.WriteLine($"Задача '{TaskName}' отмечена как выполненная");
    }
    
    public virtual void GetTaskDetails()
    {
        Console.WriteLine($"ID: {TaskId}, Задача: {TaskName}");
        Console.WriteLine($"Приоритет: {Priority}, Статус: {Status}");
        Console.WriteLine($"Исполнитель: {Assignee}, Создана: {CreatedDate.ToShortDateString()}");
    }
    
    public virtual void ReassignTo(string newAssignee)
    {
        Console.WriteLine($"Задача переназначена с {Assignee} на {newAssignee}");
        Assignee = newAssignee;
    }
    
    public void UpdatePriority(string newPriority)
    {
        Priority = newPriority;
        Console.WriteLine($"Приоритет задачи изменен на: {newPriority}");
    }
}

public class DelegateTask : Task
{
    public DateTime DueDate { get; set; }
    public string DelegatedBy { get; set; }
    
    public DelegateTask(int taskId, string taskName, string priority, string assignee, DateTime dueDate, string delegatedBy)
        : base(taskId, taskName, priority, assignee)
    {
        DueDate = dueDate;
        DelegatedBy = delegatedBy;
    }
    
    public override void MarkAsComplete()
    {
        Status = "Выполнена";
        Console.WriteLine($"Делегированная задача '{TaskName}' завершена. Срок выполнения: {DueDate.ToShortDateString()}");
    }
    
    public override void GetTaskDetails()
    {
        base.GetTaskDetails();
        Console.WriteLine($"Срок выполнения: {DueDate.ToShortDateString()}");
        Console.WriteLine($"Делегировал: {DelegatedBy}");
    }
    
    public void ExtendDeadline(int days)
    {
        DueDate = DueDate.AddDays(days);
        Console.WriteLine($"Срок выполнения продлен на {days} дней. Новый срок: {DueDate.ToShortDateString()}");
    }
}

public class TeamTask : Task
{
    public string TeamName { get; set; }
    public int TeamSize { get; set; }
    
    public TeamTask(int taskId, string taskName, string priority, string assignee, string teamName, int teamSize)
        : base(taskId, taskName, priority, assignee)
    {
        TeamName = teamName;
        TeamSize = teamSize;
    }
    
    public override void ReassignTo(string newAssignee)
    {
        Console.WriteLine($"Командная задача переназначена в команде '{TeamName}' с {Assignee} на {newAssignee}");
        Assignee = newAssignee;
    }
    
    public override void GetTaskDetails()
    {
        base.GetTaskDetails();
        Console.WriteLine($"Команда: {TeamName}, Размер команды: {TeamSize}");
    }
    
    public void ConductTeamMeeting()
    {
        Console.WriteLine($"Проведено собрание команды '{TeamName}' по задаче '{TaskName}'");
    }
}

public class ResearchTask : Task
{
    public string DataSource { get; set; }
    public string ResearchType { get; set; }
    
    public ResearchTask(int taskId, string taskName, string priority, string assignee, string dataSource, string researchType)
        : base(taskId, taskName, priority, assignee)
    {
        DataSource = dataSource;
        ResearchType = researchType;
    }
    
    public override void GetTaskDetails()
    {
        base.GetTaskDetails();
        Console.WriteLine($"Источник данных: {DataSource}");
        Console.WriteLine($"Тип исследования: {ResearchType}");
    }
    
    public void GenerateReport()
    {
        Console.WriteLine($"Сгенерирован отчет по исследованию '{TaskName}' на основе {DataSource}");
    }
}

public interface IReportable
{
    void GenerateReport();
}

public interface IAuditable
{
    void LogActivity(string activity);
}

public class AnalyticalTask : ResearchTask, IReportable, IAuditable
{
    public string AnalysisMethod { get; set; }
    
    public AnalyticalTask(int taskId, string taskName, string priority, string assignee, string dataSource, string researchType, string analysisMethod)
        : base(taskId, taskName, priority, assignee, dataSource, researchType)
    {
        AnalysisMethod = analysisMethod;
    }
    
    public void LogActivity(string activity)
    {
        Console.WriteLine($"Зафиксирована активность анализа: {activity}");
    }
    
    public override void GetTaskDetails()
    {
        base.GetTaskDetails();
        Console.WriteLine($"Метод анализа: {AnalysisMethod}");
    }
    
    public void PerformAnalysis()
    {
        Console.WriteLine($"Выполнен анализ методом: {AnalysisMethod}");
    }
}

public class Program
{
    public static void Main()
    {
        Console.WriteLine("=== ПРОСТОЕ НАСЛЕДОВАНИЕ ===");
        DelegateTask delegateTask = new DelegateTask(1, "Подготовить отчет", "Высокий", "Иван Иванов", new DateTime(2024, 12, 20), "Иван Иванов");
        delegateTask.GetTaskDetails();
        delegateTask.MarkAsComplete();
        delegateTask.ExtendDeadline(5);
        
        Console.WriteLine("\n=== СЛОЖНОЕ НАСЛЕДОВАНИЕ ===");
        TeamTask teamTask = new TeamTask(2, "Разработка модуля", "Средний", "Петр Петров", "DevTeam", 5);
        teamTask.GetTaskDetails();
        teamTask.ReassignTo("Мария Козлова");
        teamTask.ConductTeamMeeting();
        teamTask.UpdatePriority("Высокий");
        
        Console.WriteLine("\n=== МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ ===");
        ResearchTask researchTask = new ResearchTask(3, "Анализ рынка", "Низкий", "Федор Федоров", "Статистические данные", "Маркетинговое");
        researchTask.GetTaskDetails();
        researchTask.GenerateReport();
        researchTask.MarkAsComplete();
        
        Console.WriteLine("\n=== ГИБРИДНОЕ НАСЛЕДОВАНИЕ ===");
        AnalyticalTask analyticalTask = new AnalyticalTask(4, "Статистический анализ", "Высокий", "Степан Степанов", "База данных", "Научное", "Регрессионный анализ");
        analyticalTask.GetTaskDetails();
        analyticalTask.GenerateReport();
        analyticalTask.LogActivity("Обработка данных");
        analyticalTask.PerformAnalysis();
        analyticalTask.MarkAsComplete();
    }
}

Program.Main();
```
