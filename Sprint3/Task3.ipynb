Описание задачи: Создать базовый класс Task в C#, который будет представлять задачи внутри проекта. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. Требования к базовому классу Task: • Атрибуты: ID задачи (TaskId), Название задачи (TaskName), Приоритет задачи (Priority). • Методы: o MarkAsComplete(): метод для отметки задачи как выполненной. o GetTaskDetails(): метод для получения деталей задачи. o ReassignTo(): метод для переназначения задачи другому члену команды. Требования к производным классам:

ДелегатскаяЗадача (DelegateTask): Должна содержать дополнительные атрибуты, такие как Дата выполнения (DueDate). Метод MarkAsComplete() должен быть переопределен для включения даты выполнения в сообщение о завершении задачи.
КоманднаяЗадача (TeamTask): Должна содержать дополнительные атрибуты, такие как Команда (TeamName). Метод ReassignTo() должен быть переопределен для указания нового члена команды, которому будет переназначена задача.
ИсследовательскаяЗадача (ResearchTask) (если требуется третий класс): Должна содержать дополнительные атрибуты, такие как Исходные данные (DataSource). Метод GetTaskDetails() должен быть переопределен для отображения источников данных, используемых в задаче, вместе с другими деталями задачи.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```


using System;
using System.Collections.Generic;

public class BankAccount
{
    public string AccountNumber { get; set; }
    public decimal Balance { get; set; }
    public string OwnerName { get; set; }
    public DateTime CreatedDate { get; set; }
    public string Currency { get; set; }
    public bool IsActive { get; set; }
    
    public BankAccount(string accountNumber, decimal balance, string ownerName)
    {
        AccountNumber = accountNumber;
        Balance = balance;
        OwnerName = ownerName;
        CreatedDate = DateTime.Now;
        Currency = "RUB";
        IsActive = true;
    }
    
    public virtual void Deposit(decimal amount)
    {
        Balance += amount;
        Console.WriteLine($"Внесено {amount} на счет {AccountNumber}");
    }
    
    public void Deposit(decimal amount, string description)
    {
        Balance += amount;
        Console.WriteLine($"Внесено {amount} на счет {AccountNumber}. Описание: {description}");
    }
    
    public virtual void Withdraw(decimal amount)
    {
        if (amount > Balance)
            throw new Exception("Недостаточно средств");
        Balance -= amount;
        Console.WriteLine($"Снято {amount} со счета {AccountNumber}");
    }
    
    public void Withdraw(decimal amount, string purpose)
    {
        if (amount > Balance)
            throw new Exception("Недостаточно средств");
        Balance -= amount;
        Console.WriteLine($"Снято {amount} со счета {AccountNumber}. Цель: {purpose}");
    }
    
    public void ShowBalance()
    {
        Console.WriteLine($"Баланс счета {AccountNumber}: {Balance}");
    }
    
    public void ShowBalance(bool showDetails)
    {
        if (showDetails)
            Console.WriteLine($"Счет: {AccountNumber}, Баланс: {Balance}, Валюта: {Currency}, Активен: {IsActive}");
        else
            ShowBalance();
    }
    
    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Счет: {AccountNumber}, Владелец: {OwnerName}, Баланс: {Balance}");
    }
    
    public void CloseAccount()
    {
        IsActive = false;
        Console.WriteLine($"Счет {AccountNumber} закрыт");
    }
    
    public void ChangeCurrency(string newCurrency)
    {
        Currency = newCurrency;
        Console.WriteLine($"Валюта счета изменена на {newCurrency}");
    }
}

public class StudentAccount : BankAccount
{
    public string University { get; set; }
    public int Year { get; set; }
    public string StudentId { get; set; }
    public decimal MonthlyLimit { get; set; }
    
    public StudentAccount(string accountNumber, decimal balance, string ownerName, string university, int year)
        : base(accountNumber, balance, ownerName)
    {
        University = university;
        Year = year;
        StudentId = GenerateStudentId();
        MonthlyLimit = 50000;
    }
    
    private string GenerateStudentId()
    {
        return $"STU{DateTime.Now:yyyyMMddHHmmss}";
    }
    
    public void ApplyDiscount(decimal discount)
    {
        Balance += discount;
        Console.WriteLine($"Студенческая скидка {discount} применена");
    }
    
    public void ApplyDiscount(decimal discount, string reason)
    {
        Balance += discount;
        Console.WriteLine($"Студенческая скидка {discount} применена. Причина: {reason}");
    }
    
    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Университет: {University}, Курс: {Year}, ID: {StudentId}");
    }
    
    public void CheckMonthlyLimit()
    {
        Console.WriteLine($"Месячный лимит: {MonthlyLimit}");
    }
    
    public void IncreaseLimit(decimal amount)
    {
        MonthlyLimit += amount;
        Console.WriteLine($"Лимит увеличен на {amount}. Новый лимит: {MonthlyLimit}");
    }
}

public class PremiumStudentAccount : StudentAccount
{
    public decimal Cashback { get; set; }
    public int BonusPoints { get; set; }
    public string CardType { get; set; }
    
    public PremiumStudentAccount(string accountNumber, decimal balance, string ownerName, string university, int year, decimal cashback)
        : base(accountNumber, balance, ownerName, university, year)
    {
        Cashback = cashback;
        BonusPoints = 0;
        CardType = "Premium";
    }
    
    public void AddCashback()
    {
        Balance += Cashback;
        Console.WriteLine($"Добавлен кэшбэк {Cashback}");
    }
    
    public void AddCashback(decimal customCashback)
    {
        Balance += customCashback;
        Console.WriteLine($"Добавлен кэшбэк {customCashback}");
    }
    
    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Кэшбэк: {Cashback}, Бонусы: {BonusPoints}, Тип карты: {CardType}");
    }
    
    public void AddBonusPoints(int points)
    {
        BonusPoints += points;
        Console.WriteLine($"Добавлено {points} бонусных баллов");
    }
    
    public void RedeemPoints()
    {
        Balance += BonusPoints;
        Console.WriteLine($"Списано {BonusPoints} бонусных баллов");
        BonusPoints = 0;
    }
}

public interface ITransferable
{
    void Transfer(BankAccount target, decimal amount);
}

public interface ILoanable
{
    void TakeLoan(decimal amount);
}

public class BusinessAccount : BankAccount, ITransferable, ILoanable
{
    public string CompanyName { get; set; }
    public decimal LoanAmount { get; set; }
    public string BusinessType { get; set; }
    public int EmployeesCount { get; set; }
    
    public BusinessAccount(string accountNumber, decimal balance, string ownerName, string companyName)
        : base(accountNumber, balance, ownerName)
    {
        CompanyName = companyName;
        LoanAmount = 0;
        BusinessType = "ООО";
        EmployeesCount = 1;
    }
    
    public void Transfer(BankAccount target, decimal amount)
    {
        Withdraw(amount);
        target.Deposit(amount);
        Console.WriteLine($"Перевод {amount} на счет {target.AccountNumber}");
    }
    
    public void Transfer(BankAccount target, decimal amount, string description)
    {
        Withdraw(amount);
        target.Deposit(amount, description);
        Console.WriteLine($"Перевод {amount} на счет {target.AccountNumber}. Описание: {description}");
    }
    
    public void TakeLoan(decimal amount)
    {
        Balance += amount;
        LoanAmount += amount;
        Console.WriteLine($"Взят кредит {amount}, общий долг: {LoanAmount}");
    }
    
    public void TakeLoan(decimal amount, int months)
    {
        Balance += amount;
        LoanAmount += amount;
        Console.WriteLine($"Взят кредит {amount} на {months} месяцев, общий долг: {LoanAmount}");
    }
    
    public void PayLoan(decimal amount)
    {
        if (amount > LoanAmount)
            throw new Exception("Сумма платежа превышает долг");
        LoanAmount -= amount;
        Console.WriteLine($"Погашен кредит {amount}, остаток долга: {LoanAmount}");
    }
    
    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Компания: {CompanyName}, Кредит: {LoanAmount}, Тип бизнеса: {BusinessType}");
    }
    
    public void HireEmployee()
    {
        EmployeesCount++;
        Console.WriteLine($"Нанят сотрудник. Всего сотрудников: {EmployeesCount}");
    }
    
    public void ChangeBusinessType(string newType)
    {
        BusinessType = newType;
        Console.WriteLine($"Тип бизнеса изменен на: {newType}");
    }
}

public class AccountCollection<T> where T : BankAccount
{
    private List<T> _accounts = new List<T>();
    
    public void AddAccount(T account)
    {
        _accounts.Add(account);
        Console.WriteLine($"Счет {account.AccountNumber} добавлен в коллекцию");
    }
    
    public void RemoveAccount(T account)
    {
        _accounts.Remove(account);
        Console.WriteLine($"Счет {account.AccountNumber} удален из коллекции");
    }
    
    public void DisplayAllAccounts()
    {
        Console.WriteLine($"\n=== ВСЕ СЧЕТА В КОЛЛЕКЦИИ ({_accounts.Count}) ===");
        foreach (var account in _accounts)
        {
            account.DisplayInfo();
        }
    }
    
    public T FindAccount(string accountNumber)
    {
        return _accounts.Find(acc => acc.AccountNumber == accountNumber);
    }
}

{
    {
        Console.WriteLine("=== ПРОСТОЕ НАСЛЕДОВАНИЕ ===");
        var student = new StudentAccount("STU001", 1000, "Иван", "МГУ", 2);
        student.DisplayInfo();
        student.ApplyDiscount(100);
        student.ApplyDiscount(50, "акция");
        student.ShowBalance(true);
        student.IncreaseLimit(10000);
        
        Console.WriteLine("\n=== СЛОЖНОЕ НАСЛЕДОВАНИЕ ===");
        var premiumStudent = new PremiumStudentAccount("PSTU001", 2000, "Мария", "МГУ", 3, 50);
        premiumStudent.DisplayInfo();
        premiumStudent.AddCashback();
        premiumStudent.AddCashback(25);
        premiumStudent.AddBonusPoints(100);
        premiumStudent.ShowBalance();
        
        Console.WriteLine("\n=== МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ ===");
        var business = new BusinessAccount("BUS001", 5000, "Петр", "ООО Ромашка");
        business.DisplayInfo();
        business.TakeLoan(1000);
        business.TakeLoan(2000, 12);
        business.Transfer(student, 500);
        business.Transfer(premiumStudent, 300, "оплата услуг");
        business.PayLoan(300);
        business.HireEmployee();
        business.DisplayInfo();
        
        Console.WriteLine("\n=== GENERIC КЛАССЫ ===");
        var accountCollection = new AccountCollection<BankAccount>();
        accountCollection.AddAccount(student);
        accountCollection.AddAccount(premiumStudent);
        accountCollection.AddAccount(business);
        accountCollection.DisplayAllAccounts();
        
        var foundAccount = accountCollection.FindAccount("STU001");
        if (foundAccount != null)
        {
            Console.WriteLine($"\nНайден счет: {foundAccount.AccountNumber}");
            foundAccount.ShowBalance();
        }
    }
}
```
